```python
import pandas as pd
import matplotlib.pyplot as plt
from simpsons_utils import *
```

```python
raw_data = pd.read_csv(SORTED_INFO)
```

```python
male_data = raw_data[raw_data["gender"]=="m"]
female_data = raw_data[raw_data["gender"]=="f"]

print("Male Characters:")
print(male_data["word_count"].describe())
print()
print("Female Characters:")
print(female_data["word_count"].describe())

plt.plot(range(len(male_data["word_count"].unique())), male_data["word_count"].value_counts())
plt.xlabel("Word Count Per Line")
plt.ylabel("Number of Lines")
plt.title("Male Character Line Word Counts")
plt.show()

plt.plot(range(len(female_data["word_count"].unique())), female_data["word_count"].value_counts())
plt.xlabel("Word Count Per Line")
plt.ylabel("Number of Lines")
plt.title("Male Character Line Word Counts")
plt.show()
```
