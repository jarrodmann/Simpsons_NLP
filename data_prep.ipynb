```python
from simpsons_utils import *
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
raw_script_lines = pd.read_csv(SCRIPT_FILE)
filtered_script_lines = raw_script_lines[raw_script_lines["speaking_line"]==True]
filtered_script_lines = filtered_script_lines[KEY_EPISODE_INFORMATION]
sorted_script_lines = filtered_script_lines.sort_values(by=EPISODE_SORTING_KEY)
sorted_script_lines
```

```python
raw_character_file = pd.read_csv(CHARACTER_FILE)
character_file = raw_character_file.drop(["name", "normalized_name"], axis=1)
labeled_script_lines = pd.merge(sorted_script_lines, character_file, left_on="character_id", right_on="id", how="left")
labeled_script_lines = labeled_script_lines.drop("id", axis=1)
labeled_script_lines = labeled_script_lines.dropna(axis=0)
labeled_script_lines.to_csv(SORTED_INFO, index=False)
```

```python
labeled_script_lines
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')

stops = set(stopwords.words('english'))
print(stopwords.words('english'))
```

```python
print(labeled_script_lines.shape)
```

```python
# Remove stop words
stop_words_data = labeled_script_lines.copy()


i = 0
for line in stop_words_data["spoken_words"]:

    tokens = word_tokenize(line)
    filtered_list = [word.lower() for word in tokens if not word.lower() in stops and word.isalpha()]
    filtered_sentence = ' '.join(filtered_list)
    stop_words_data.iloc[i, 3] = filtered_sentence
    i += 1
    
    if i % 1000 == 0:
        print(f"Completed {i} / {stop_words_data.shape[0]}")
stop_words_data = stop_words_data.dropna(axis=0)
```
